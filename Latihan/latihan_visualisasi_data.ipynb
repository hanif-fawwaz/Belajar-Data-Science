```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# latihan 1
bulan = ["Jan", "Feb", "Mar", "Apr", "Mei", "Juni"]
penjualan_toko_a = [50, 100, 145, 200, 111, 66]
penjualan_toko_b = [60, 180, 155, 189, 111, 66]
```

```python
# Toko A
plt.plot(bulan, penjualan_toko_a, marker='o', color='blue', linestyle='-')
plt.title("Penjualan Bulanan Toko A")
plt.xlabel("Bulan")
plt.ylabel("Jumlah Penjualan")
plt.grid(True)
plt.show()
```

```python
# Toko B
plt.plot(bulan, penjualan_toko_b, marker='o', color='blue', linestyle='-')
plt.title("Penjualan Bulanan Toko B")
plt.xlabel("Bulan")
plt.ylabel("Jumlah Penjualan")
plt.grid(True)
plt.show()
```

```python
# Latihan 2
data = {
    "Nama": ["Hanif", "Arib", "Fawwaz", "Cici", "Dewi", "Eko", "Fajar"],
    "Nilai": [85, 70, 90, 65, 88, 92, 75],
    "Kelas": ["A","B", "A", "C", "B", "A", "C"]
}
df = pd.DataFrame(data)
```

```python
sns.histplot(df["Nilai"], bins=5, kde=True)
plt.title("Nilai Rata-Rata Siswa")
plt.show()
```
