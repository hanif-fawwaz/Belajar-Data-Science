Kenapa Visusalisasi?

Dalam Data Science, data seringkali besar dan sulit dipahami kalu cuma angka.
Dengan Grafik, kita bisa cepat tren, pola, otlier. 

```python
# Visualisasi dengan menggunakan Matplotlib
import matplotlib.pyplot as plt
```

```python
# Data Contoh
bulan = ["Jan", "Feb", "Mar", "Apr", "Mei"]
penjualan = [100, 150, 120, 180, 200]
```

```python
# Membuat grafik garis
plt.plot(bulan, penjualan, marker='o', color='blue', linestyle='-')
plt.title("Grafik Penjualan Bulanan")
plt.xlabel("Bulan")
plt.ylabel("Jumlah Penjualan")
plt.grid(True)
plt.show()
```

penjelasan:
- plt.plot() : membuat grafik garis
- marker='o' : titik pada tiap data
- plt.grid(True) : menampilkan garis bantu

```python
# Visualisasi dengan Bar chart
plt.bar(bulan,penjualan, color='orange')
plt.title("Penjualan Bulanan")
plt.xlabel("Bulan")
plt.ylabel("Jumlah penjualan")
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Visualisasi dengan menggunakan seaborn
# data nilai siswa
data = {
    "Nama": ["Hanif", "Arib", "Fawwaz", "Cici", "Dewi", "Eko", "Fajar"],
    "Nilai": [85, 70, 90, 65, 88, 92, 75],
    "Kelas": ["A","B", "A", "C", "B", "A", "C"]
}
```

```python
df = pd.DataFrame(data)
```

```python
sns.histplot(df["Nilai"], bins=5, kde=True)
plt.title("Distribusi Nilai Siswa")
plt.show()
```
